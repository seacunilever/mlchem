```python
from mlchem.chem.manipulation import (MolGenerator,
                                      MolCleaner)

# Load your list of SMILES
# Here, dataset is generated for demonstrative purposes
generator = MolGenerator()
generator.generate_smiles(template_smiles='CCCN',
                          n_molecules=100,
                          n_fragments=5,
                          n_substitutions=10,
                          attempt_limit=5000,
                          include_extremities=True)

# Show the generated SMILES retain the initial template
display(generator.smiles_generated[:5])

# Perform all cleaning steps
cleaner = MolCleaner(input_smiles_list=generator.smiles_generated)
cleaner.full_clean()
```
